```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

df = pd.read_csv('../data/titanic.csv')
df.head()
```

```python
sns.countplot(x='Sex', data=df);
```

```python
sns.catplot(x='Survived', col='Sex', kind='count', data=df);
```

```python
df["AgeCohort"] = ((df["Age"].fillna(df["Age"].median()) / 5).round(0) * 5).astype(pd.Int64Dtype())
sns.catplot(x='AgeCohort', col='Survived', data=df, kind="count", hue='Survived')
```

```python
df["FareCohort"] = ((df["Fare"].fillna(df["Fare"].median()) / 5).round(0) * 5).astype(pd.Int64Dtype())
sns.catplot(x='FareCohort', col='Survived', data=df, kind="count", hue='Survived')
```

```python
sns.catplot(x='FareCohort', row="AgeCohort",col='Survived', data=df, kind="count", hue='Survived')
```

```python
from sklearn.metrics import accuracy_score

df['model_1'] = 0
accuracy_score(df['Survived'], df['model_1'])
```

```python
df['model_2'] = df.Sex == 'female'
accuracy_score(df['Survived'], df['model_2'])
```
